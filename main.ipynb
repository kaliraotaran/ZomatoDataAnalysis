```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data .csv')
df.head()
```

```python
#split the values of rate so we are only left with the first value

def  split_rate(rate):
    value = str(rate).split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(split_rate)
print (df.head())
```

```python
sns.histplot(df['listed_in(type)'], )
plt.xlabel('Type of Restaurant')
```

```python
sns.lineplot(x ='listed_in(type)', y= 'votes', data=df, )
plt.xlabel('Types of Restaurants', c='blue', size=15)
plt.ylabel('Votes', c='blue', size = 15) 
```

```python
sns.countplot(x = df['online_order'])
```

```python
mul_people = df['approx_cost(for two people)']
sns.countplot(x = mul_people)
# this plots the approx spent by 2 people like ~6 people spent 100 on a 2-person dinner
```

```python
table = df.pivot_table(index = 'listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(table, annot=True, cmap='YlGnBu', fmt= 'd')
plt.xlabel('Online Order')
plt.ylabel('Listed In (type)')
plt.show()
```
